```python
covid_data = covid[covid['source_type'] == 'agences-regionales-sante'].reset_index()
```

Check if info are unique for each item

```python
for dt in covid_data['date'].unique():
    for dep in covid_data['maille_nom'].unique():
        if covid_data[(covid_data['date'] == dt) & (covid_data['maille_nom'] == dep)].shape[0] > 1:
            print(covid_data[(covid_data['date'] == dt) & (covid_data['maille_nom'] == dep)])
            ind = covid_data[(covid_data['date'] == dt) & (covid_data['maille_nom'] == dep)].index
            print(ind)
```

We want to build a new dataframe having for each date all the departments and their Covid status

```python
#Let's collect all the dept 
covid_fr_dep['maille_nom'].unique()
dep_list = covid_fr_dep['maille_nom'].unique().tolist()
```

```python
dep_list
```

```python
#il manque la Réunion
dep_list.append('La Réunion')
```

```python
#flist all the days from the begining till the previous day
date_list = pd.date_range(start = '2020-01-24', end = pd.datetime.today()-timedelta(1)).tolist()
```

```python
len(dep_list), len(date_list)
```

Our new dataframe should have (number of days*101 rows = 6161)

```python
## add all the departments for each date
for dt in date_list:
    if covid_fr_dep[covid_fr_dep['date'] == dt].shape[0] != len(dep_list):
        existing_dep = covid_fr_dep[covid_fr_dep['date'] == dt]['maille_nom']
        non_existing_dep = np.setdiff1d(dep_list,existing_dep)
        for nom_dep in non_existing_dep:
            covid_fr_dep = covid_fr_dep.append({'date': dt, 'granularite': 'departement', 'maille_nom': nom_dep, 'cas_confirmes': 0.0, 'deces': 0.0, 'reanimation': 0.0, 'hospitalises': 0.0, 'gueris':0.0}, ignore_index = True)
```
